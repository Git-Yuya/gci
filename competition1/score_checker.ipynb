```python
import numpy as np
import pandas as pd

from sklearn.metrics import accuracy_score
```

```python
# ファイルをセッションストレージにアップロード
test = pd.read_csv("test_Perished.csv")
submission = pd.read_csv("submission.csv")
```

```python
# テストデータのPerished
y = test["Perished"].values
# 提出ファイルのPerished
pred = submission["Perished"].values

print(accuracy_score(y, pred))
```
